```python
#librsries to be used
import random
# List of words with more than six letters
word_list = ["butterfly", "elephant", "paraben", "vitamin", "photoageing","quantum"]

# Randomly select a word from the list
selected_word = random.choice(word_list)
print(f"Randomly selected word: {selected_word}")

# Initialize the hidden word with underscores
hidden_word = ["_"] * len(selected_word)

# Set the number of allowed incorrect guesses
guess_limit = 3

# To track the number of incorrect guesses
incorrect_guesses = 0

Unused_Letters=[]
Valid_Letters=[] #To ensure the player guesses a letter instead of special character
for i in "abcdefghijklmnopqrstuvwxyz":
    Unused_Letters.append(i)
    Valid_Letters.append(i)

# While the number of incorrect guesses is less than the limit and the word is not fully guessed
while incorrect_guesses < guess_limit and "_" in hidden_word:
    # Display the current state of the word (underscores and guessed letters)
    print("\n" + " ".join(hidden_word))
    print(f"Remaining guesses: {guess_limit - incorrect_guesses}")

    # Prompt the user to guess a letter
    current_guess = input("Guess a letter: ").lower()

    # Validate the guessed letter
    if current_guess in Unused_Letters:
        Unused_Letters.remove(current_guess)
        
        # Check if the letter is in the selected word
        if current_guess in selected_word:
            # Update the hidden word with the correctly guessed letter
            for idx, letter in enumerate(selected_word):
                if letter == current_guess:
                    hidden_word[idx] = current_guess
        else:
            # Increment the number of incorrect guesses
            incorrect_guesses += 1
            print(f"Incorrect guess! {current_guess} is not in the word.")
    else:
        if current_guess not in Valid_Letters:
            print("Invalid. Please try again.")
        else:
            print("You already guessed that letter")

    # Display the updated game state
    print("____________________________________________________\n")
    print(" ".join(hidden_word))
    print(f"Remaining guesses: {guess_limit - incorrect_guesses}")

# Check if the player has won or lost
if "_" not in hidden_word:
    print("\nCongratulations! You guessed the word correctly.")
else:
    print(f"\nGame over! The word was '{selected_word}'. Better luck next time.")
```
