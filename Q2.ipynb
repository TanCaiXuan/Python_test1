```python
# libraries to  be used
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("done")
```

```python
# Read the dataset into a DataFrame
spotify = pd.read_csv("spotify.csv")

spotify.info()
```

```python
# to change the date from STr to date
if 'track_album_release_date' in spotify.columns: 
    spotify['track_album_release_date'] = pd.to_datetime(spotify['track_album_release_date'], errors='coerce')
    
# Drop rows with NaT in 'track_album_release_date' after conversion
spotify = spotify.dropna(subset=['track_album_release_date'])

spotify
```

```python
# Filter for tracks released in the year 2017
spotify_2017 = spotify[spotify['track_album_release_date'].dt.year == 2017]
spotify_2017 
```

```python
# Group by genre and calculate the required statistics
genre_stats = spotify_2017.groupby('playlist_genre').agg(
    average_loudness=('loudness', 'mean'),
    mode=('mode', lambda x: x.mode()[0]),  # Mode might return multiple values
    average_speechiness=('speechiness', 'mean'),
    average_acousticness=('acousticness', 'mean'),
    average_instrumentalness=('instrumentalness', 'mean'),
    average_liveness=('liveness', 'mean'),
    average_valence=('valence', 'mean'),
    average_tempo=('tempo', 'mean'),
    average_duration_minutes=('duration_ms', lambda x: x.mean() / 60000)  # Convert ms to minutes
).reset_index()

genre_stats
```

```python
# Save the findings into a CSV file
genre_stats.to_csv("genre_statistics_2017.csv", index=False)
```

```python
# Calculate the correlation between danceability and energy
correlation = spotify_2017['danceability'].corr(spotify_2017['energy'])

print(f"Correlation between danceability and energy: {correlation:.2f}")

print("It means there is a weak correlation")
print("It may not be useful to consider these two features as directly influencing each other in terms of music track characteristics")
```

```python
# Further analysis of other factors influencing danceability and energy
correlation_matrix = spotify_2017[['danceability', 'energy', 'loudness', 'speechiness', 
                               'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Correlation Matrix of Features Influencing Danceability and Energy")
plt.tight_layout()
plt.savefig("correlation_matrix.png")
plt.show()
```

```python
print("1. Coefficient < 0 (Negative Correlation) means one variable increases, the other tends to decrease significantly.\n2. Correlation Coefficient = 0 (No Correlation) means there is no systematic pattern in how the variables relate to each other.\n3. Correlation Coefficient > 0 (Positive Correlation) means one variable increases, the other also tends to increase significantly.")
```

```python
from sklearn.linear_model import LinearRegression

# Assuming you have filtered the DataFrame for 2017 tracks
# Define features and target variable
features = spotify_2017[['danceability', 'energy', 'acousticness']]
target = spotify_2017['track_popularity']  # Adjust to your actual popularity column name

# Fit linear regression model
reg = LinearRegression()
reg.fit(features, target)

# Feature names
feature_names = ['danceability', 'energy', 'acousticness']

# Print coefficients one by one
for feature, coef in zip(feature_names, reg.coef_):
    print(f"Coefficient of '{feature}': {coef}")
    
print("\nWith a positive coefficient, danceability increases popularity when raised.\nSimilarly for  acousticness.\nMeanwhile, higher energy decreases popularity.\n")

print("Intercept:", reg.intercept_)
# Optional: Predict popularity for the dataset and evaluate the model
predictions = reg.predict(features)

# Calculate R^2 score to evaluate the model
r_squared = reg.score(features, target)
print("\nR^2 Score:", r_squared)
print(f"The R-squared of {round(r_squared,4)} indicates our audio features explain {round(r_squared*100,2)}% of the variance in track popularity. ")

```
